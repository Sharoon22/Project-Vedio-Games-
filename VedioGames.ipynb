## Sales by Region


```python
regions=game[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
sbr=regions.sum()
```

```python
sbr
```

```python
plt.figure(figsize=(6,6))
sbr.plot(kind='bar', color=['red', 'Blue', 'Green', 'orange'])
plt.title('Sales By Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.show()
```

## Top selling Geners

```python
topg=game.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8, 5))
topg.plot(kind='bar', color='green')
plt.title('Top selling Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.xticks(rotation=90)
plt.show()
```

## Top selling Platforms

```python
topplat=game.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
topplat.plot(kind='bar', color='Green')
plt.title('Top Selling Platforms by Global Sales')
plt.xlabel('Platform')
plt.ylabel('Global Sales')
plt.xticks(rotation=90)
plt.show()
```

## Top 10 Games By Global

```python
top10=game[['Name','Global_Sales']].sort_values(by='Global_Sales', ascending=False).head(10)
```

```python
top10
```

```python
sns.lineplot(x='Name',y='Global_Sales',data=top10,marker='o')
plt.xticks(rotation=90)
plt.show()
```

## Number of games by genere

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=game, x='Genre', palette='Set1')
plt.title('Number of Games by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.xticks(rotation=90)
plt.show()
```
